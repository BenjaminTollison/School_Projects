```python
####                                Fixing Case 3                                 ####

# Starting with Maplesoft I evaluated the assumed solution of Ae^alphax + Be^-alphax #
#  By using the appropriate boundary conditions I derivated a new solution for u(x)  #
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,1000)
alpha = np.array([0.25, 0.50,0.73, 1., 2., 4.,6.33, 8., 10.73])
def u_c(alpha,x):
    r = 0.1
    h = 1
    P = np.pi * r**2
    u_x = (50 * h * P * np.exp(-alpha)) / (alpha * np.cosh(alpha)**2) * np.sinh(alpha * x) + 100 / np.cosh(alpha) * np.cosh(alpha * x)
    return u_x

#### graphing
for i in alpha:
    plt.plot(x, u_c(i,x), label='' + str(i) + ' \u03B1')
plt.title('Case C exact Temperature')
plt.xlabel('x')
plt.ylabel('u(x) in Celcius')
plt.legend()
plt.show()
```

```python
def Q_c(alpha, x):
    r = 0.1
    h = 1
    P = np.pi * r**2
    k = 1
    A = P
    u_prime = (50 * h * P * np.exp(-alpha)) / (np.cosh(alpha)**2) * np.cosh(alpha * x) - (100 * alpha) / np.cosh(alpha) * np.sinh(alpha * x)
    Q = k*A * u_prime
    return Q
#### graphing
for i in alpha:
    plt.plot(x, Q_c(i,x), label='' + str(i) + ' \u03B1')
plt.title('Case C exact Bulk Heat Flux')
plt.xlabel('x')
plt.ylabel('Q(x) in Celcius')
plt.legend()
plt.show()
```
