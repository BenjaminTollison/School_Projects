```python
### AOA 6 ###
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
AOA_6_data = open('Neg6.txt', 'r')
reader = csv.reader(AOA_6_data, delimiter='\t')
time = []
P_0 = []
P_1 = []
for row in reader:
    time.append(float(row[0]))
    P_0.append(float(row[1]))
    P_1.append(float(row[2]))
Mach = []
for i in range(len(time)):
    gamma = 1.4
    Mach.append(np.sqrt((2*((P_0[i]/P_1[i])**((gamma-1)/gamma) - 1)) / (gamma-1)))
plt.plot(time,P_0,'b',label='P_0')
plt.plot(time,P_1,'r',label='P_1')
plt.plot(time,Mach,'g',label='Mach')
plt.legend()
plt.xlabel('Time (s)')
AOA_6_data.close()
output_text_file = open("Output_{}.txt".format('Neg6.txt'), "w")
output_text_file.write("Time\tP_0\tP_1\n"+
                       "{}\n{}\n{}\n".format(time,P_0,P_1) +
                       "")
output_text_file.close()
    
```
