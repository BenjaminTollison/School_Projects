```python
import sympy as sp
from IPython.display import display, Math, Latex
from sympy.solvers.solveset import linsolve
from sympy import  lambdify, Matrix 
import numpy as np
import matplotlib.pyplot as plt
def displayEquations(LHS,RHS):
    left = sp.latex(LHS)
    right = sp.latex(RHS)
    display(Math(left + '=' + right))
```

```python
# 2
    # a
R = 2
r1_values = np.linspace(0, R, 100)
r2_values = np.linspace(R, 2*R , 100)
C = 1
v_theta = [lambda r: (C*r) / R**2,lambda r: C/r]
plt.plot(r1_values, v_theta[0](r1_values))
plt.plot(r2_values, v_theta[1](r2_values))
plt.title('v_theta(r)')
plt.xlabel('r')
plt.ylabel('v_theta')
plt.show()
```

```python
# 3 
    # a
A,r,theta = sp.symbols('A r theta')
psi = A * r**(7/5) * sp.sin((7/5)*(sp.pi - theta))
v_r, v_theta = 1/r * psi.diff(theta), - psi.diff(r)
displayEquations('v_r',v_r)
displayEquations('v_theta',v_theta)
```

```python
eqn1 = v_theta.subs({theta: sp.pi})
eqn2 = v_r.subs({theta: sp.pi})
```

```python
# Graphing 3
A = 1
psi_np = lambda r, theta: A * r**(7/5) * np.sin((7/5)*(np.pi - theta))
v_r_np = lambda r, theta: -1.4*A*r**(2/5)*np.cos((7/5)*(theta - np.pi))
v_theta_np = lambda r, theta: 1.4*A*r**(2/5)*np.sin((7/5)*(theta - np.pi))
r_values = np.linspace(0,10,100)
a = np.arange(0,10,1)
theta_values = np.linspace(0,2*np.pi,100)
for i in a:
    plt.plot(theta_values,psi_np(i,theta_values),label='r =' + str(i))
plt.title('Psi(r,theta)')
plt.xlabel('theta')
plt.ylabel('Psi')
plt.legend()
plt.show()
```

```python
for i in a:
    plt.plot(theta_values,v_r_np(i,theta_values),label='r =' + str(i))
plt.title('v_r(r,theta)')
plt.xlabel('theta')
plt.ylabel('v_r')
plt.legend()
plt.show()
```

```python
for i in a:
    plt.plot(theta_values,v_theta_np(i,theta_values),label='r =' + str(i))
plt.title('v_theta(r,theta)')
plt.xlabel('theta')
plt.ylabel('v_theta')
plt.legend()
plt.show()
```

```python
# finding the psi stag
eqn1 = (7/5)*(sp.pi - theta)
eqn2 = (7/5)*(sp.pi - theta) - sp.pi
eqn3 = (7/5)*(sp.pi - theta) - 2*sp.pi
eqn4 = (7/5)*(sp.pi - theta) + 2*sp.pi
eqn5 = (7/5)*(sp.pi - theta) + sp.pi
zero1 = sp.solve(eqn1)
zero2 = sp.solve(eqn2)
zero3 = sp.solve(eqn3)
zero4 = sp.solve(eqn4)
zero5 = sp.solve(eqn5)
displayEquations('zero1', zero1)
displayEquations('zero2', zero2)
displayEquations('zero3', zero3)
displayEquations('zero4', zero4)
displayEquations('zero5', zero5)
```
